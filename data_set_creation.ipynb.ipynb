```python
import pandas as pd 
import numpy as np 
import random 
```

```python
states = ['Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 'Goa', 'Gujarat', 'Haryana',
          'Himachal Pradesh', 'Jharkhand', 'Karnataka', 'Kerala', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 
          'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 
          'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal', 'Andaman and Nicobar Islands', 
          'Chandigarh', 'Dadra and Nagar Haveli and Daman and Diu', 'Delhi', 'Lakshadweep', 'Puducherry']
genders = ['Male', 'Female']
insulins = ['Lyumjev', 'Humalog', 'Novolog']
feedback_options = ['Positive', 'Neutral', 'Negative']
years = [2020, 2021, 2022, 2023, 2024]
def generate_name():
    first_names = ['John', 'Priya', 'Ravi', 'Sarah', 'Ali', 'Sanjay', 'Rina', 'Anil', 'Zara', 'Vikram']
    last_names = ['Sharma', 'Patel', 'Singh', 'Khan', 'Gupta', 'Das', 'Bose', 'Mehta', 'Kapoor', 'Malik']
    return f"{random.choice(first_names)} {random.choice(last_names)}"
```

```python
def get_prescription_volume(feedback):
    if feedback == 'Positive':
        return random.randint(50, 100)  
    elif feedback == 'Neutral':
        return random.randint(20, 50)
    else:
        return random.randint(5, 20)
```

```python
size = 5000
```

```python
data = {
    'id': [i for i in range(1, size+1)],
    'patient_name': [generate_name() for _ in range(size)],
    'state': [random.choice(states) for _ in range(size)],
    'gender': [random.choice(genders) for _ in range(size)],
    'age': [random.randint(18, 80) for _ in range(size)],
    'insulin_name': [random.choice(insulins) for _ in range(size)],
    'feedback': [random.choice(feedback_options) for _ in range(size)],
    'prescription_volume': [get_prescription_volume(fb) for fb in [random.choice(feedback_options) for _ in range(size)]],
    'year': [random.choice(years) for _ in range(size)]
}
```

```python
df = pd.DataFrame(data)
```
